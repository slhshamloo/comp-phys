## Conclusion

# Percolation Probability

```julia
len, minprob, maxprob, samples, runs = 100, 0.55, 0.64, 50, 10000
```

## Full

```julia
probs = collect(range(0, 1, length=samples))
percolate_avg = zeros(samples)

for run in 1:runs
    randgrid = rand(len, len)
    for sample in 1:samples
        grid = randgrid .<= probs[sample]
        percolate = Percolation.dfspercolate(grid)
        percolate_avg[sample] += percolate
    end
end

percolate_avg ./= runs
percolate_σ = .√(percolate_avg - percolate_avg .^ 2)

# save data
open("percolate-data/full-$len.txt", "w") do io
    writedlm(io, [probs percolate_avg percolate_σ])
end
```

```julia
# load data from previous runs
data = readdlm("percolate-data/full-$len.txt")
probs, percolate_avg, percolate_σ = data[:, 1], data[:, 2], data[:, 3]
```

```julia
p = scatter(probs, percolate_avg, yerr=percolate_σ, markerstrokecolor=:red,
    color=:mediumblue, lw=2, legend=false, fontfamily="Computer Modern",
    title=L"L=%$len,\ \textrm{averaged\ over\ %$runs \ runs}", xlabel=L"p", ylabel=L"Q")
plot!(p, probs, percolate_avg, color=:mediumblue, lw=2)

savefig(p, figpath * "percolate-full-$len.pdf")
display(p)
```

## Zoom

```julia
probs = collect(range(minprob, maxprob, length=samples))
percolate_avg = zeros(samples)

for run in 1:runs
    randgrid = rand(len, len)
    for sample in 1:samples
        grid = randgrid .<= probs[sample]
        percolate = Percolation.dfspercolate(grid)
        percolate_avg[sample] += percolate
    end
end

percolate_avg ./= runs
percolate_σ = .√(percolate_avg - percolate_avg .^ 2)

# save data
open("percolate-data/zoom-$len.txt", "w") do io
    writedlm(io, [probs percolate_avg percolate_σ])
end
```

```julia
p = plot(probs, percolate_avg, ribbon=percolate_σ, fillcolor=:lightskyblue,
    color=:mediumblue, marker=:circle, lw=2, legend=false, fontfamily="Computer Modern",
    title=L"L=%$len,\ \textrm{averaged\ over\ %$runs \ runs}", xlabel=L"p", ylabel=L"Q")

savefig(p, figpath * "percolate-zoom-$len.pdf")
display(p)
```

# $Q_\infty$

```julia
len, minprob, samples, runs = 200, 0.5, 50, 1000
area = len^2
```

```julia
probs = collect(range(minprob, 1, length=samples))
q∞_avg, q∞_σ = zeros(samples), zeros(samples)

for run in 1:runs
    randgrid = rand(len, len)
    for sample in 1:samples
        grid = randgrid .<= probs[sample]
        clustersize, _, _, = Percolation.hkpercolate(grid; qinfty=true)
        q∞ = clustersize / area
        q∞_avg[sample] += q∞
        q∞_σ[sample] += q∞ ^ 2
    end
end

q∞_avg ./= runs
# absolute value added to ensure no negative numbers resulting from floating-point error at small σ
q∞_σ = .√(abs.(q∞_σ ./ runs - q∞_avg .^ 2))

# save data
open("qinfty-data/qinfty-$len.txt", "w") do io
    writedlm(io, [probs q∞_avg q∞_σ])
end
```

```julia
# load data from previous runs
data = readdlm("qinfty-data/qinfty-$len.txt")
probs, q∞_avg, q∞_σ = data[:, 1], data[:, 2], data[:, 3]
```

```julia
p = scatter(probs, q∞_avg, yerr=q∞_σ, markerstrokecolor=:red,
    color=:mediumblue, lw=2, legend=false, fontfamily="Computer Modern",
    title=L"L=%$len,\ \textrm{averaged\ over\ %$runs \ runs}", xlabel=L"p", ylabel=L"Q_\infty")
plot!(p, probs, q∞_avg, color=:mediumblue, lw=2)

savefig(p, figpath * "percolate-qinfty-$len.pdf")
display(p)
```

# Diffusion-Limited Aggregation
## Packages

```julia
include("RandomWalk.jl")
using Plots, ColorSchemes, LaTeXStrings, DelimitedFiles
import PyPlot as plt
```

## Configurations

```julia
figpath = "fig/"
rcParams = plt.PyDict(plt.matplotlib["rcParams"])
rcParams["text.usetex"] = true
```

## Visualization

```julia
width, height = 500, 300
surface = falses(height, width)
surface[1, :] .= true
surfaces = [copy(RandomWalk.dlaline!(surface, 1000)) for i in 1:7]

maxheight = height
while true ∉ surfaces[7][maxheight - 1, :]
    maxheight -= 1
end

fig, ax = plt.subplots()
ax.set_xlim(0, width)
ax.set_ylim(0, maxheight)

rainbow = plt.get_cmap("rainbow")
for i in 7:-1:1
    ax.imshow(surfaces[i], cmap=plt.ColorMap("cust", vcat([1 1 1 0],
        collect(rainbow((i - 1) / 6))')),  interpolation="nearest")
end

display(fig)
plt.savefig(figpath * "dla-$width.pdf", bbox_inches="tight")
```

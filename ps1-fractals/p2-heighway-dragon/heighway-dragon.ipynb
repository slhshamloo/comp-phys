# Heighway Dragon
## Imports

```python
from dragon import heighway_dragon
from matplotlib import rcParams, pyplot as plt

rcParams['text.usetex'] = True
directory = "../fig/"
```

## Plot
### Steps

```python
def plot(ax, i):
    x, y = heighway_dragon((0, 1), (0, 0), i)

    ax.plot(x[:len(x)//2 + 1], y[:len(y)//2 + 1], 'b')
    ax.plot(x[len(x)//2:], y[len(y)//2:], 'r')

    ax.set_axis_off()
    ax.set_aspect("equal")
    ax.set_title("Step " + str(i))
```

```python
fig, axs = plt.subplots(3, 3)
fig.tight_layout()

axs = axs.flatten()
for i in range(9):
    axs[i].set_aspect("equal")
    plot(axs[i], i + 1)

plt.savefig(directory + "heighway.pdf", bbox_inches="tight")
```

### Large

```python
fig, ax = plt.subplots()

x, y = heighway_dragon((0, 1), (0, 0), 16)
plt.plot(x[:len(x)//2 + 1], y[:len(y)//2 + 1], 'b', lw=0.3)
plt.plot(x[len(x)//2:], y[len(y)//2:], 'r', lw=0.3)

ax.set_axis_off()
plt.savefig("../fig/heighway-large.pdf", bbox_inches = "tight")
```

# Koch Snowflake

## Imports and Config

```python
from koch import koch
from matplotlib import rcParams, pyplot as plt

rcParams['text.usetex'] = True
directory = "../fig/"
```

## Koch Curve
### Steps

```python
fig, axs = plt.subplots(2, 3, figsize=(6.4, 2))
fig.tight_layout()

for i in range(6):
    x, y = koch((0, 1), (0, 0), i)

    ax = axs.flat[i]
    ax.plot(x, y, 'k', lw=0.5)

    ax.set_axis_off()
    ax.set_aspect("equal")
    ax.set_title("Step " + str(i))

# plt.savefig(directory + "koch.pdf", bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(8, 4 / 3**0.5))

x, y = koch((0, 1), (0, 0))
ax.plot(x, y, 'k', lw=0.3)

ax.set_axis_off()
# plt.savefig(directory + "koch-large.pdf", bbox_inches="tight")
```

## Snowflake

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

x, y = koch((0, 0.5, 1, 0), (0, 3**0.5 / 2, 0, 0))
ax.plot(x, y, 'k', lw=0.3)

ax.set_axis_off()
# plt.savefig(directory + 'koch-tri.pdf', bbox_inches='tight')
```

## Rainbow Coloring Each Step
### Curve

```python
fig, ax = plt.subplots(figsize=(8, 4 / 3**0.5))
ax.set_aspect("equal")
cmap = plt.get_cmap("rainbow")

for i in range(7):
    x, y = koch((0, 1), (0, 0), i)
    ax.plot(x, y, lw=0.3, c=cmap(i / 6), label=f"step {i}")

ax.set_axis_off()
ax.legend(fontsize=8.5)
# plt.savefig(directory + 'koch-color.pdf', bbox_inches='tight')
```

### Snowflake

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")
cmap = plt.get_cmap("rainbow")

for i in range(6, -1, -1):
    x, y = koch((0, 0.5, 1, 0), (0, 3**0.5 / 2, 0, 0), i)
    ax.fill(x, y, lw=0, c=cmap(i / 6), label=f"step {i}")

ax.set_axis_off()
ax.legend(fontsize=8.5)
# plt.savefig(directory + 'koch-tri-color.pdf', bbox_inches='tight')
```

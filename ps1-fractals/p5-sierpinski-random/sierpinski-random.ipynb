# Randomly Generating the Sierpinski Triangle
## Imports and config

```python
from sierpinski import sierpinski_random
from matplotlib import rcParams, pyplot as plt

rcParams['text.usetex'] = True
directory = "../fig/"
```

## Plots
### Steps

```python
fig, axs = plt.subplots(2, 2, figsize=(5.5, 2.75 * 3**0.5))
for ax in axs.flat:
    ax.set_axis_off()

axs[0, 0].plot(*sierpinski_random((0, 1), (0, 1), 10, 1000), 'b,')
axs[0, 1].plot(*sierpinski_random((0, 1), (0, 1), 10, 2500), 'b,')
axs[1, 0].plot(*sierpinski_random((0, 1), (0, 1), 10, 5000), 'b,')
axs[1, 1].plot(*sierpinski_random((0, 1), (0, 1), 10, 10000), 'b,')

axs[0, 0].set_title("1000 Samples")
axs[0, 1].set_title("2500 Samples")
axs[1, 0].set_title("5000 Samples")
axs[1, 1].set_title("10000 Samples")

plt.savefig(directory + "sierpinski-random.png", bbox_inches="tight", dpi=100)
```

### Large

```python
fig, ax = plt.subplots(figsize=(5.5, 2.75 * 3**0.5))
ax.set_aspect('auto')
ax.set_axis_off()

x, y = sierpinski_random((0, 1), (0, 1), 10, 50000)
ax.plot(x, y, 'b,')
ax.set_title("50000 Samples")

plt.savefig(directory + "sierpinski-random-large.png", bbox_inches="tight", dpi=100)
```

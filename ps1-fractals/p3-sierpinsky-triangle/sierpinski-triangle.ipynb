# Sierpi≈Ñski Triangle
## Imports and Config

```python
from sierpinski import sierpinski
from fractal import get_polygons
from matplotlib import rcParams, pyplot as plt

rcParams['text.usetex'] = True
directory = "../fig/"
```

## Plot
### Steps

```python
fig, axs = plt.subplots(2, 3)

for i in range(6):
    x, y = sierpinski((0, 0.5, 1), (0, 3**0.5 / 2, 0), i)

    ax = axs.flat[i]
    ax.add_collection(get_polygons(x, y, color='b'))

    ax.set_axis_off()
    ax.set_aspect("equal")
    ax.autoscale_view(tight=True)
    ax.set_title("Step " + str(i))

# plt.savefig(directory + "sierpinski.pdf", bbox_inches="tight")
```

### Large

```python
fig, ax = plt.subplots(figsize=(8, 4 * 3**0.5))

x, y = sierpinski((0, 0.5, 1), (0, 3**0.5 / 2, 0), 10)
ax.add_collection(get_polygons(x, y, color=plt.get_cmap("prism")))

ax.set_axis_off()
ax.autoscale_view(tight=True)
# plt.savefig(directory + "sierpinski-large.pdf", bbox_inches="tight")
```

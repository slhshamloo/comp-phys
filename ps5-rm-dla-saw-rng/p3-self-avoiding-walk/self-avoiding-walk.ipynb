# Self-Avoiding Random Walk
## Packages and Config

```julia
include("SelfAvoidingWalk.jl")
using Plots, LaTeXStrings, DelimitedFiles

figpath = "../fig/"
Plots.default(fontfamily="Computer Modern", linewidth=3, markersize=5, legend=false)
```

## Path Count

```julia
pathcount = [countsaw(n) for n in 1:20]

# save data to file
open("saw-path-counts.txt", "w") do io
    writedlm(io, [1:20 pathcount])
end
```

```julia
# load data from file
pathcount = readdlm("saw-path-counts.txt")[:, 2]
```

```julia
p = plot(pathcount, marker=:circle, c=:blue, title="Self-Avoiding Walks",
    xlabel="Path Length", ylabel="Number of Paths")

savefig(p, figpath * "saw-path-count.pdf")
display(p)
```

```julia
p = plot(pathcount[1:7], marker=:circle, c=:blue, title="Self-Avoiding Walks",
    xlabel="Path Length", ylabel="Number of Paths")

savefig(p, figpath * "saw-path-count-small.pdf")
display(p)
```

## Compare Against Random Walk

```julia
p = plot([2n for n in 1:20], marker=:circle, c=:red, legend=:topleft,
    label="Random Walk", xlabel="Path Length", ylabel="ln(Number of Paths)")
plot!(p, log.(pathcount), marker=:circle, c=:blue, label="Self-Avoiding Walk")

savefig(p, figpath * "self-avoiding-vs-random-walk.pdf")
display(p)
```

```julia
p = plot(pathcount ./ [4^n for n in 1:20], marker=:circle, c=:blueviolet,
    title="Ratio of Self-Avoiding to Random\nWalk Path Count",
    xlabel="Path Length", ylabel="Ratio")

savefig(p, figpath * "self-avoiding-to-random-walk-ratio.pdf")
display(p)
```

```julia
p = plot(log.(pathcount ./ [4^n for n in 1:20]), marker=:circle, c=:blueviolet,
    title="Ratio of Self-Avoiding to Random\nWalk Path Count",
    xlabel="Path Length", ylabel="ln(Ratio)")

savefig(p, figpath * "self-avoiding-to-random-walk-ratio-log.pdf")
display(p)
```

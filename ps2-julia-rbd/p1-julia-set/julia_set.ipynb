# Julia Set
## Imports and Config

```python
from julia_set import julia_quad, crop_julia
from custom_colors import wbrykmap
from matplotlib.colors import LogNorm
from matplotlib import rcParams, pyplot as plt

directory = "../fig/julia/"
rcParams['text.usetex'] = True
```

## Quadratic Polynomials $f(z) = z^2 + c$
### Initialization

```python
plt.tight_layout()
fig, axs = plt.subplots(3, 2, figsize=(8, 10))
axs = axs.flatten()
for ax in axs:
    ax.set_axis_off()

fractals = (
    crop_julia(julia_quad(-0.6)),
    crop_julia(julia_quad(-0.123 + 0.745j)), # douady's rabbit fractal
    crop_julia(julia_quad(-0.391 - 0.587j)), # siegel disk fractal
    crop_julia(julia_quad(-1j)), # the dendrite fractal
    crop_julia(julia_quad(-0.4 + 0.6j)),
        # golder ratio - 2 + (golden ratio - 1)i = -0.4 + 0.6i
    crop_julia(julia_quad(-0.8 + 0.16j))
)

titles = (r"$c = 0.6$", r"$c = -0.123 + 0.745i$", r"$c = -0.391 - 0.587i$",
          r"$c = -i$", r"$c = -0.4 + 0.6i$", r"$c = -0.8 + 0.16i$")

for i in range(3):
    axs[i].set_title(titles[i])
    axs[i].imshow(fractals[i], origin="lower", interpolation="none",
                  norm=LogNorm(), cmap="inferno")
for i in range(3, 6):
    axs[i].set_title(titles[i])
    axs[i].imshow(fractals[i], origin="lower", interpolation="none",
                  cmap=wbrykmap())

plt.savefig(directory + "julia-main.pdf", bbox_inches="tight")
```
